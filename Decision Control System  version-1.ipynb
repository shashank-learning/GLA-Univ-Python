# Decision Making

What is an if statement?
- An if statement tests for a condition, and then responds to that condition. 
- If the condition is true, then whatever action is listed next gets carried out. 
- You can test for multiple conditions at the same time, and respond appropriately to each condition.

# Flow Chart If-Else


```python
if 10>20 :
    print("greater")
    print("i am in if")
else :
    print("I am outside if")
```

```python
if -10:
    print("hi")
else:
    print('byee')
```

# The if-elif...else chain
You can test whatever series of conditions you want to, and you can test your conditions in any combination you
want.


```python
a=10
b=200000
c=30
if a>b and a>c :
    print("Greatest no is {}".format(a))
elif b>c :
    print("Greatest no is {}".format(b))
else :
    print("Greatest no is {}".format(c))
```

# Random Numbers in Python

Python defines a set of functions that are used to generate or manipulate random numbers

randint() is an inbuilt function of the random module in Python3.It gives random number in between beg and end.

<b>Syntax :</b>
 > randint(start, end)
 
 > (start, end) : Both of them must be integer type values.
 

```python
import random

n=random.randint(100,99999)
print(n)
n=random.randint(100,99999)
print(n)
n=random.randint(100,99999)
print(n)
```
